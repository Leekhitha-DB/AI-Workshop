```python
import numpy as np

import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
data = np.random.randn(1000)

# Line plot
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title('Line Plot')

# Bar chart
plt.subplot(2, 3, 2)
plt.bar(categories, values)
plt.title('Bar Chart')

# Histogram
plt.subplot(2, 3, 3)
plt.hist(data, bins=30)
plt.title('Histogram')

# Scatter plot
plt.subplot(2, 3, 4)
plt.scatter(x, y)
plt.title('Scatter Plot')

# Pie chart
plt.subplot(2, 3, 5)
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart')

# Box plot
plt.subplot(2, 3, 6)
plt.boxplot(data)
plt.title('Box Plot')

plt.tight_layout()
plt.show()
```
