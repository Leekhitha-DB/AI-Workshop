```python
def classify_sentence(sentence):
    sentence = sentence.strip()  # remove leading/trailing spaces
    
    if sentence.endswith('?'):
        return "Question"
    elif sentence.endswith('!'):
        return "Exclamation"
    else:
        return "Statement"

# Example usage
examples = [
    "How are you?",
    "This is amazing!",
    "I am learning Python."
]

for s in examples:
    print(f"'{s}' -> {classify_sentence(s)}")
```
