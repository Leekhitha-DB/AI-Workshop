```python
from collections import Counter
import re

def text_to_word_frequency(text):
    """
    Convert text to a frequency count of words.
    
    Args:
        text (str): Input text to analyze
        
    Returns:
        Counter: Dictionary-like object with words and their frequencies
    """
    # Convert to lowercase and split into words
    words = re.findall(r'\b\w+\b', text.lower())
    
    # Count word frequencies
    word_freq = Counter(words)
    
    return word_freq


# Example usage
sample_text = "Hello world hello python python python"
frequencies = text_to_word_frequency(sample_text)
print(frequencies)
```
