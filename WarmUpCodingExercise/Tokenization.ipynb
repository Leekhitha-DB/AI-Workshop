```python
# Simple text tokenization in Python

# Sample text
text = "Hello there! This is a simple example. Let's break text into smaller units."

# --- Breaking into words ---
words = text.split()  # splits by whitespace
print("Words:", words)

# --- Breaking into sentences ---
# Using nltk for sentence tokenization
import nltk
nltk.download('punkt')  # download tokenizer models (only needed once)
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print("Sentences:", sentences)
```
