```python
import string

def jaccard_similarity(sentence1, sentence2):
    # Helper function to clean and tokenize
    def tokenize(sentence):
        translator = str.maketrans('', '', string.punctuation)
        cleaned = sentence.translate(translator).lower()
        return set(cleaned.split())
    
    # Tokenize both sentences
    set1 = tokenize(sentence1)
    set2 = tokenize(sentence2)
    
    # Compute Jaccard similarity
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    
    if not union:  # avoid division by zero
        return 0.0
    
    similarity = len(intersection) / len(union)
    return similarity

# Example usage
s1 = "The cat sat on the mat."
s2 = "A cat is sitting on the mat."
print("Jaccard Similarity:", jaccard_similarity(s1, s2))
```
