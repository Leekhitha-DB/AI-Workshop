```python
import string
from collections import Counter

# Define a basic list of stopwords (you can expand this or use nltk's stopwords)
stop_words = {
    "the", "is", "in", "and", "to", "a", "of", "on", "for", "with", "at", "by", 
    "an", "be", "this", "that", "it", "as", "are", "was", "were", "from"
}

def extract_keywords(text, top_n=5):
    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator).lower()
    
    # Tokenize into words
    words = cleaned_text.split()
    
    # Filter out stop words
    meaningful_words = [word for word in words if word not in stop_words]
    
    # Count word frequencies
    freq = Counter(meaningful_words)
    
    # Return top N keywords
    return freq.most_common(top_n)

# Example usage
text = "The cat sat on the mat and the dog barked loudly at the cat."
keywords = extract_keywords(text, top_n=3)
print("Top Keywords:", keywords)
```
