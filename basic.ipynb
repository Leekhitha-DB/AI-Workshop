```python
# Examples of data types and variables in Python

# Integer
x = 5
print(f"x is {x}, type: {type(x)}")

# Float
y = 3.14
print(f"y is {y}, type: {type(y)}")

# String
name = "Alice"
print(f"name is {name}, type: {type(name)}")

# Boolean
is_true = True
print(f"is_true is {is_true}, type: {type(is_true)}")

# List
my_list = [1, 2, 3]
print(f"my_list is {my_list}, type: {type(my_list)}")

# Dictionary
my_dict = {"key": "value"}
print(f"my_dict is {my_dict}, type: {type(my_dict)}")
```

```python
# Example of control flow using if statements with existing variables

if x > 3:
    print(f"x ({x}) is greater than 3")
else:
    print(f"x ({x}) is not greater than 3")

if is_true:
    print("The boolean is_true is True")
else:
    print("The boolean is_true is False")

if len(my_list) > 2:
    print(f"my_list has more than 2 elements: {my_list}")
else:
    print(f"my_list has 2 or fewer elements: {my_list}")
```

```python
# Example of control flow using if-else with existing variables

if y > 3:
    print(f"y ({y}) is greater than 3")
else:
    print(f"y ({y}) is not greater than 3")

if name == "Alice":
    print(f"Hello, {name}!")
else:
    print(f"Hello, stranger!")

if "key" in my_dict:
    print(f"my_dict contains 'key': {my_dict['key']}")
else:
    print("my_dict does not contain 'key'")
```

```python
if x % 2 == 0:
    print(f"x ({x}) is even")
else:
    print(f"x ({x}) is odd")

if y < 3.5:
    print(f"y ({y}) is less than 3.5")
else:
    print(f"y ({y}) is 3.5 or greater")

if name.startswith("A"):
    print(f"name ({name}) starts with 'A'")
else:
    print(f"name ({name}) does not start with 'A'")

if len(my_list) == 3:
    print(f"my_list has exactly 3 elements: {my_list}")
else:
    print(f"my_list does not have exactly 3 elements: {my_list}")

if is_true and "key" in my_dict:
    print(f"is_true is {is_true} and my_dict has 'key': {my_dict['key']}")
else:
    print("Condition not met")
```

```python
# Example of control flow using for loops with existing variables

for item in my_list:
    print(f"Item in my_list: {item}")

for key, value in my_dict.items():
    print(f"Key: {key}, Value: {value}")

for i in range(x):
    print(f"Loop iteration: {i}")
```

```python
# Example of control flow using while loops with existing variables

counter = 0
while counter < len(my_list):
    print(f"Counter: {counter}, Item: {my_list[counter]}")
    counter += 1

# While loop with a condition based on existing variable
num = x
while num > 0:
    print(f"Counting down: {num}")
    num -= 1

# While loop that modifies a variable
count = 0
while count < 3 and is_true:
    print(f"Count: {count}, is_true: {is_true}")
    count += 1
```
